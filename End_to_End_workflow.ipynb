```python
!pip install graphviz

import graphviz

# Create a new directed graph
dot = graphviz.Digraph(comment='System Workflow')
dot.attr(rankdir='TB')

# Add nodes
dot.node('User', 'User\nüôã')
dot.node('InputPrompt', 'Input Text Prompting\nüó®Ô∏è')
dot.node('LLM', 'LLM Processing\nüß†')
dot.node('Validation', 'Validation and Error Checking\n‚úÖ')
dot.node('ScriptGen', 'Script Generation\nüìú')
dot.node('Robot', 'Robot Action Execution\nü§ñ')

# Add edges
dot.edge('User', 'InputPrompt', 'Clean the living room')
dot.edge('InputPrompt', 'LLM', 'Process natural language')
dot.edge('LLM', 'Validation', 'Generate command structure')
dot.edge('Validation', 'ScriptGen', 'Validated command')
dot.edge('ScriptGen', 'Robot', 'Execute cleaning routine')
dot.edge('Robot', 'User', 'Cleaning complete')

# Render the graph
dot.render('system_workflow', format='png', cleanup=True)
dot
```
